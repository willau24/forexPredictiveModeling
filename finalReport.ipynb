```python
#Determine trade validity

country_to_code = {"United Kingdom": "GBP", "Japan": "JPY", "Australia": "AUD", "EU": "EUR", "Canada": "CAD"}

for country, code in country_to_code.items():
    if code + "/USD" in next_day_returns_dict:
        if next_day_returns_dict[code + "/USD"] > 0:
            print(f"\nUSD is appreciating compared to {country}")
            print("Confidence Level: 1")
            
            if country in country_gdp_comparison:
                if country_gdp_comparison[country]:
                    print(f"{country} is growing faster than the US. USD is therefore depreciating")
                else:
                    print(f"{country} is growing slower than the US. USD is therefore appreciating")
                    print("Confidence Level: 2")
                
                    if country in country_inflation_comparison:
                        if country_inflation_comparison[country]:
                            print(f"Higher inflation in {country} indicates appreciation of USD in exchange")
                            print("Confidence Level: 3")
                            
                            if code in country_sentiment:
                                if country_sentiment[code]:
                                    print(f"There is positive sentiment around {code}/USD exchange")
                                    print("Confidence Level: 4")
                                else:
                                    print(f"There is negative sentiment around {code}/USD exchange")
                                    
                        else:
                            print(f"Higher inflation in {country} indicates depreciation of USD in exchange")
                        
        else:
            print(f"\nUSD is depreciating compared to {country}")
            
            if country in country_gdp_comparison:
                if country_gdp_comparison[country]:
                    print(f"{country} is growing faster than the US. USD is therefore depreciating")
                    print("Confidence Level: 2")
                    
                    if country in country_inflation_comparison:
                        if country_inflation_comparison[country]:
                            print(f"Higher inflation in {country} indicates appreciation of USD in exchange")
                            
                        else:
                            print(f"Higher inflation in {country} indicates depreciation of USD in exchange")
                            print("Confidence Level: 3")
                            
                            if code in country_sentiment:
                                if country_sentiment[code]:
                                    print(f"There is positive sentiment around {code}/USD exchange")
                                    print("Confidence Level: 4")
                                else:
                                    print(f"There is negative sentiment around {code}/USD exchange")
                            
                else:
                    print(f"{country} is growing slower than the US. USD is therefore appreciating")
```
